```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from pathlib import Path
```

```python
plt.style.use('ggplot')
path = Path('..', 'data', 'raw')
credit = pd.read_csv(path / 'credit_data.csv').rename(columns=str.lower)
cat_names = ['home', 'marital', 'records', 'job']
credit[cat_names] = credit[cat_names].astype('category')
```

```python
for id, col in enumerate(cat_names):
    plt.subplot(2, 2, id + 1)
    cross_tab = (pd.crosstab(credit[col], credit['status'], normalize=True)
        .reset_index()
        .pipe(pd.melt, id_vars=col)
        .assign(value = lambda df: df['value'] * 100))
    ax = sns.barplot(x=col, y='value', hue='status', data=cross_tab)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
fig = plt.gcf()
fig.set_size_inches(25.5, 10.5)
fig.suptitle('Variáveis categóricas', fontsize=20)
plt.savefig('../reports/figures/categorical.png')
```
